## Transformar Dados

```python
# Concatenar tabelas
df = pd.concat([streamings1, streamings2, streamings3], axis=0)
df
```

```python
df.shape
```

```python
# Renomear as colunas
novo_nome_colunas = {'endTime': 'Data da reprodução',
                     'artistName': 'Artista',
                     'trackName': 'Música',
                     'msPlayed': 'Milissegundos de Reprodução'}
df = df.rename(columns=novo_nome_colunas)
```

## Limpeza de dados

```python
# Somando as linhas sem valores
df.isnull().sum()
```

```python
df.info()
```

```python
# Converter coluna 'Data da reprodução' para o tipo datetime
df['Data da reprodução'] = pd.to_datetime(df['Data da reprodução'])
```

```python
# Conferir quais são a primeira e última data de registro
registro1 = df['Data da reprodução'].min() 
ultimo_registro = df['Data da reprodução'].max()

print(f'O primeiro registro dessa playlist foi: {registro1}')
print(f'O último registro dessa playlist foi: {ultimo_registro}')
```

## Análise de Dados

```python
# Qtde de artistas que aparecem nessa playlist
valores_unicos = df['Artista'].unique()
print(len(valores_unicos))
```

```python
# Calcular quantas vezes o mesmo artista aparece - Top 10
contagem_artista = df['Artista'].value_counts().head(10)
contagem_artista
```

```python
# Qtde de músicas diferentes que aparecem nessa playlist
valores_unicos = df['Música'].unique()
print(len(valores_unicos))
```

```python
# Calcular qtas vezes a mesma Música foi executada - Top 15
contagem_musica = df['Música'].value_counts()
contagem_musica.head(15)
```

```python
# Mapear os valores numericos dos dias da semana para pt-Br
dia_semana_map = {
    0: 'Segunda-feira',
    1: 'Terça-feira',
    2: 'Quarta-feira',
    3: 'Quinta-feira',
    4: 'Sexta-feira',
    5: 'Sábado',
    6: 'Domingo'
}

# extraindo o dia da semana
df['Dia_Semana'] = df['Data da reprodução'].dt.dayofweek.map(dia_semana_map)
df
```

```python
# Contar o número de execuções de música por dia da semana
contagem_dias = df['Dia_Semana'].value_counts()
contagem_dias
```

## Criar uma playlist com 15 músicas de acordo com o dia da semana

```python
#concatenar artista e música
df['artista_musica'] = df['Artista'] + ' - ' + df['Música']
df.head(10)
```

```python
# Quantidade de artista-músicas diferentes
valores_unicos = df['artista_musica'].unique()
print(len(valores_unicos))
```

```python
#contar os valores únicos
recomendacao = df['artista_musica'].value_counts()
recomendacao
```

```python
# Filtra as músicas populares que têm 10 ou mais aparições
musicas_populares = recomendacao[recomendacao >= 10]
musicas_populares.tail(15)
```

```python
#Quantidade de linhas artista-músicas
musicas_populares.shape
```

```python
#Valores da coluna Dia_Semana
df['Dia_Semana'].unique()
```

```python
#Filtrar as músicas de acordo com o dia da semana 
dia_selecionado = 'Terça-feira'
musicas_do_dia = df[(df['Dia_Semana'] == dia_selecionado) & 
                    (df['artista_musica'].isin(musicas_populares.index))]

#Embaralhar as músicas do dia para tornar a seleção aleatória
musicas_do_dia = musicas_do_dia.sample(frac=1) #todas as linhas

#Selecionar as primeiras 15 músicas como recomendação
playlist_recomendada = musicas_do_dia.head(15)

#Configuração para exibir o texto completo das colunas
pd.set_option('display.max_colwidth', None) 

#Exibir resultado
display(playlist_recomendada[['artista_musica']])
```
